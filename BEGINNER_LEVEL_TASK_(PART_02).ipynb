```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("/B Task2 dataset.csv")
```

```python
df.head(4)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
daily_sales = df.groupby('date')['sales'].sum().reset_index()
daily_sales
```

```python
daily_sales['weekly_ma'] = daily_sales['sales'].rolling(window=7).mean()
```

```python
daily_sales['monthly_ma'] = daily_sales['sales'].rolling(window=30).mean()
```

```python
plt.figure(figsize=(12,6))
plt.plot(daily_sales['date'], daily_sales['sales'], label='Actual Sales')
plt.plot(daily_sales['date'], daily_sales['weekly_ma'], label='Weekly Moving Avg')
plt.plot(daily_sales['date'], daily_sales['monthly_ma'], label='Monthly Moving Avg')

plt.legend()
plt.title("Sales Forecast using Moving Average")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```
